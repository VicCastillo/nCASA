```python
#extract information from wav file and use specgram to transform
def deconstruct_WAV(wav):
    samplerate = wav[0]
    signal = wav[1]
    samples = len(signal)
    time = samples/samplerate
    print "Sample Rate = ", samplerate, "Hz"
    print "Total time = ", time, "sec"
    print "specgram returns spectral image:"
    specArray, freqs, bins, im  = specgram(signal, cmap=cm.gist_heat)
    #maxpower = np.amax(specArray)
    #specArray /= maxpower
    #return maxpower, specArray, samplerate, time
    return specArray, freqs, bins, im, time
    
newA = deconstruct_WAV(cardinal)
```

```python
#show freq power data vs time as an array
def showarray(data):
    specArray = data[0]
    time = data[4]
    maxpower = np.amax(specArray)
    plt.figure(figsize=(15,10))
    plt.xlabel("Time(sec)")
    plt.ylabel("Freq(Hz)")
    plt.axis([0, len(data[2]), 0, len(data[1])])
    xlocs,xlabs = plt.xticks()
    plt.xticks(xlocs,xlocs.__mul__(time/len(data[2])).round(2))
    ylocs,ylabs = plt.yticks()
    plt.yticks(ylocs,ylocs.__mul__(440/14).round(2))
    plt.pcolor(specArray.__div__(maxpower))
    plt.colorbar()
    print "Color plot of spectral-temporal data matrix: "


showarray(newA)
```

```python
import scipy, pylab

def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X

def istft(X, fs, T, hop):
    x = scipy.zeros(T*fs)
    framesamp = X.shape[1]
    hopsamp = int(hop*fs)
    for n,i in enumerate(range(0, len(x)-framesamp, hopsamp)):
        x[i:i+framesamp] += scipy.real(scipy.ifft(X[n]))
    return x


#f0 = 440         # Compute the STFT of a 440 Hz sinusoid
fs = 11025        # sampled at 8 kHz
T = 3.0            # lasting 5 seconds
framesz = 0.50  # with a frame size of 50 milliseconds
hop = 0.020      # and hop size of 20 milliseconds.

# Create test signal and STFT.
t = scipy.linspace(0, T, T*fs, endpoint=False)
#x = scipy.sin(2*scipy.pi*f0*t)
#X = stft(x, fs, framesz, hop)

X = stft(cardinal[1], fs, framesz, hop)

# Plot the magnitude spectrogram.
pylab.figure()
pylab.imshow(sc-ipy.absolute(X.T), origin='lower', aspect='auto',
             interpolation='nearest')
pylab.xlabel('Time')
pylab.ylabel('Frequency')
pylab.show()

# Compute the ISTFT.
xhat = istft(X, fs, T, hop)

# Plot the input and output signals over 0.1 seconds.
T1 = int(0.1*fs)

pylab.figure()
pylab.plot(t[:T1], xhat[:T1], t[:T1], xhat[:T1])
pylab.xlabel('Time (seconds)')

pylab.figure()
pylab.plot(t[-T1:], xhat[-T1:], t[-T1:], xhat[-T1:])
pylab.xlabel('Time (seconds)')

```
