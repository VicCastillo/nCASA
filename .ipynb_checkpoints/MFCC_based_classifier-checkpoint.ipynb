```python
#import scipy, pylab

def stft(x, fs, framesamp, hopsamp):
    w = sp.hamming(framesamp)
    X = sp.array([sp.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X

def istft(X, fs, T, hopsamp):
    x = sp.zeros(T*fs)
    framesamp = 128 #X.shape[1]
    for n,i in enumerate(range(0, len(x)-framesamp, hopsamp)):
        x[i:i+framesamp] += sp.real(sp.ifft(X[n]))
    return x

def plot_time(signal):
    plt.figure(figsize=(20,3))
    plt.plot(signal)
    plt.xlabel("Samples")
    plt.show()

    ##Need to redo scaling for tick labels
def colorarray(specArray):
    maxpower = np.amax(specArray)
    plt.figure(figsize=(15,10))
    plt.xlabel("Time(sec)")
    plt.ylabel("Freq(Hz)")

    #xlocs,xlabs = plt.xticks() 
    #ylocs,ylabs = plt.yticks()
    #plt.yticks(ylocs,ylocs.__mul__(440/14).round(2))
    plt.pcolor(specArray.__div__(maxpower))
    plt.colorbar()
    print "Color plot of spectral-temporal data matrix: "
```

```python
fs = wavedata[0] #sample rate
x  = wavedata[1] #time series
plot_time(x)

X = stft(x, fs, 128, 64) #transform into frequency domain

##Squaring real and imaginary components and taking square root

Xnew = np.sqrt(X.real**2 + X.imag**2)
colorarray(Xnew.T)

nsamples = 40879
T = nsamples/fs
xnew = istft(X, fs, T, 64) #transfer back ingto time domain

xnewint = xnew.astype(np.int16)
plot_time(xnewint)

print X.shape
```

```python
###Good explanation of MFCCs here
##http://www.practicalcryptography.com/miscellaneous/machine-learning/guide-mel-frequency-cepstral-coefficients-mfccs/

###This is how MFCCs work

###Take a signal and split it up into 25ms samples
###Use a mel filterbank to seperate the sample into frequency components
###The components near 0Hz are seperated out well, higher frequencies less so
##This is because the human ear can seperate lower frequencies better than higher ones

###Sum all of the energies at each filter level to get an idea of the power spectrum envelope for that frame
###Take the log of these to normalize the energies (since they go up by a factor of 10 per decibel intensity)
##Take the cosine transform to decorrelate the energies at every level
```

```python
##now we are going to implement the gaussian mixture model for this dataset


##shows seperation of Gaussian distribution by overlaying ellipses on graph
##from http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_classifier.html#example-mixture-plot-gmm-classifier-py
# Author: Ron Weiss <ronweiss@gmail.com>, Gael Varoquaux
# License: BSD 3 clause


def make_ellipses(gmm, ax):
    for n, color in enumerate('rgb'):
        v, w = np.linalg.eigh(gmm._get_covars()[n][:2, :2])
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v *= 9
        ell = mpl.patches.Ellipse(gmm.means_[n, :2], v[0], v[1],
                                  180 + angle, color=color)
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)

wav_data = zero130[:-1] ##data from the wave
wav_num_samples = zero130.shape[0] ##number of mfcc samples for training data


wav_data_labels_neg = np.array([0] * int(wav_num_samples/2)) ##test run, splitting half of one file
wav_data_labels_pos = np.array([1] * int(wav_num_samples/2))

wav_data_labels = np.concatenate([wav_data_labels_neg,wav_data_labels_pos]) ##adding the labels together
##assuming indices of data and labels are sames

##splitting the data into training and testing data
strat = StratifiedKFold(wav_data_labels, n_folds = 4)
wav_label_train, wav_label_test = next(iter(strat)) ##randomly generated indices of 75% training and 25% test samples
```

```python
print wav_data.shape ##371 samples of 13 mfccs

X_train = wav_data[wav_label_train] ##training data with proper indices from stratified k-fold. [0] because it returns an array
y_train = wav_data_labels[wav_label_train]##training labels with proper indices

X_test = wav_data[wav_label_test]##testing data with proper indices
y_test = wav_data_labels[wav_label_test]##testing labels with proper indices
```

```python
# Try GMMs using different types of covariances.
classifiers = dict((covar_type, GMM(n_components=n_classes,
                    covariance_type=covar_type, init_params='wc', n_iter=20))
                   for covar_type in ['spherical', 'diag', 'tied', 'full'])

n_classifiers = len(classifiers)

plt.figure(figsize=(3 * n_classifiers / 2, 6))
plt.subplots_adjust(bottom=.01, top=0.95, hspace=.15, wspace=.05,
                    left=.01, right=.99)


for index, (name, classifier) in enumerate(classifiers.items()):
    # Since we have class labels for the training data, we can
    # initialize the GMM parameters in a supervised manner.
    classifier.means_ = np.array([X_train[y_train == i].mean(axis=0) for i in xrange(n_classes)])

    # Train the other parameters using the EM algorithm.
    classifier.fit(X_train)

    h = plt.subplot(2, n_classifiers / 2, index + 1)
    #make_ellipses(classifier, h)

    for n, color in enumerate('rgb'):
        data = wav_data[wav_data_labels == n]
        plt.scatter(data[:, 0], data[:, 1], 0.8, color=color,
                    label="")
    # Plot the test data with crosses
    for n, color in enumerate('rgb'):
        data = X_test[y_test == n]
        plt.plot(data[:, 0], data[:, 1], 'x', color=color)

    y_train_pred = classifier.predict(X_train)
    train_accuracy = np.mean(y_train_pred.ravel() == y_train.ravel()) * 100
    plt.text(0.05, 0.9, 'Train accuracy: %.1f' % train_accuracy,
             transform=h.transAxes)

    y_test_pred = classifier.predict(X_test)
    test_accuracy = np.mean(y_test_pred.ravel() == y_test.ravel()) * 100
    plt.text(0.05, 0.8, 'Test accuracy: %.1f' % test_accuracy,
             transform=h.transAxes)

    plt.xticks(())
    plt.yticks(())
    plt.title(name)

plt.legend(loc='lower right', prop=dict(size=12))


plt.show()
```

```python
##need to learn about LDA/QDA and how it relates to Gaussian

##spherical GMM
sGMM = GMM(n_components = n_classes, covariance_type = 'spherical')##spherical GMM
sGMM.fit(X_train)##fit the GMM with training data
###will compute its own distribution
p_results = sGMM.predict(X_test)##test how well the distribution reacts


# Compute confusion matrix
confusion_m = confusion_matrix(y_test, p_results)

#plot
plt.matshow(confusion_m)
plt.title('Confusion matrix between predicted and actual')
plt.colorbar()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
